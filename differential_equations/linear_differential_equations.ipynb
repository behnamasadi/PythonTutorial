# Linear differential equations
Linear differential equations are a type of differential equation where the **unknown function** and its **derivatives** $ y, y^{'}, y^{''}, y^{'''},... $ appear to the power of **1**. They are linear in the sense that the **function and its derivatives** are to the **first power** (not squared, cubed, etc.) and are not multiplied together.


$ a_n(x)\frac{d^n y}{dx^n} + a_{n-1}(x)\frac{d^{n-1} y}{dx^{n-1}} + \ldots + a_1(x)\frac{dy}{dx} + a_0(x)y = g(x) $

where $ a_n(x), a_{n-1}(x), \ldots, a_1(x), a_0(x) $ are coefficients which can be functions of $ x $ but not $ y $, and $ g(x) $ is the non-homogeneous term.

### Example:

Consider the linear differential equation:

$ a(x) y'' + b(x) y' + c(x) y = g(x) $

where $ y'' $ denotes the second derivative of $ y $ with respect to $ x $, $ y' $ denotes the first derivative of $ y $ with respect to $ x $, and $ a(x) $, $ b(x) $, $ c(x) $, and $ g(x) $ are given functions of $ x $. This is a second order linear differential equation.

### Explanation:

In the generic formula given above:

- $ a(x) y'' $ represents the term involving the second derivative of the function $ y $.
- $ b(x) y' $ represents the term involving the first derivative of the function $ y $.
- $ c(x) y $ represents the term involving the function $ y $ itself.
- $ g(x) $ on the right-hand side makes this a non-homogeneous linear differential equation. If $ g(x) $ were zero, it would be a homogeneous linear differential equation.


There are several different types of linear differential equations, which can be categorized based on their order, homogeneity, and the number of variables involved.

# 1. **Ordinary Linear Differential Equations (ODEs):**
   These are differential equations that contain one independent variable and one or more of its derivatives with respect to that variable.

   - **First-Order Linear ODE:** The general form is $\frac{dy}{dx} + P(x)y = Q(x) $, where $P(x) $ and $Q(x) $ are functions of $x $.
   - **Second-Order Linear ODE:** The general form is $\frac{d^2y}{dx^2} + P(x)\frac{dy}{dx} + Q(x)y = R(x) $, where $P(x) $, $Q(x) $, and $R(x) $ are functions of $x $.
   - **Higher-Order Linear ODE:** For $n $-th order, the general form is $\frac{d^ny}{dx^n} + a_{n-1}(x)\frac{d^{n-1}y}{dx^{n-1}} + \ldots + a_1(x)\frac{dy}{dx} + a_0(x)y = f(x) $, where the coefficients $a_i(x) $ are functions of $x $.

### Example Comparisons:
- **First-Order Linear**: $ \frac{dy}{dx} + p(x)y = q(x) $ (Both first-order and linear)
- **First-Order Non-Linear**: $ \frac{dy}{dx} = y^2 + x $ (First-order but not linear)
- **Second-Order Linear**: $ \frac{d^2 y}{dx^2} + p(x)\frac{dy}{dx} + q(x)y = r(x) $ (Linear but not first-order)


# 2. **Linear Partial Differential Equations (PDEs):**
   These involve partial derivatives of a function that depends on several independent variables.

   - **First-Order Linear PDE:** An example would be $a(x,y)\frac{\partial u}{\partial x} + b(x,y)\frac{\partial u}{\partial y} = c(x,y) $, where $u $ is a function of $x $ and $y $.
   - **Second-Order Linear PDE:** A common form is $a(x,y)\frac{\partial^2 u}{\partial x^2} + 2b(x,y)\frac{\partial^2 u}{\partial x \partial y} + c(x,y)\frac{\partial^2 u}{\partial y^2} + \ldots = f(x,y) $.

# 3. **Homogeneous vs. Non-Homogeneous:**
   Linear differential equations can also be classified based on whether they are homogeneous or non-homogeneous.

   - **Homogeneous Linear Differential Equation:** All terms involve the function or its derivatives. In other words, the equation is set to 0: $a_n(x)\frac{d^ny}{dx^n} + \ldots + a_1(x)\frac{dy}{dx} + a_0(x)y = 0 $.
   - **Non-Homogeneous Linear Differential Equation:** There is an additional term that is not a function of the unknown function or its derivatives (i.e., it's not set to 0): $a_n(x)\frac{d^ny}{dx^n} + \ldots + a_1(x)\frac{dy}{dx} + a_0(x)y = f(x) $, where $f(x) \neq 0 $.

# 4. **Systems of Linear Differential Equations:**
   These are sets of linear differential equations involving multiple functions and their derivatives. They are often written in matrix form.

Each type of linear differential equation has its own methods for finding solutions. For ODEs, these might include techniques like integrating factors, characteristic equations, variation of parameters, or Laplace transforms. For PDEs, methods might involve separation of variables, Fourier transforms, or the method of characteristics.
