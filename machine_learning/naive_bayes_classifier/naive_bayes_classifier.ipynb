## Bayesian Inference
Bayesian inference is a method of statistical inference in which Bayes' theorem is used to update the probability for a hypothesis as more evidence or information becomes available. It's a powerful framework for understanding how our beliefs or states of knowledge about the world should logically change in the light of new evidence. Here's a breakdown of its components and how it works:

1. **Prior Probability $P(H)$**: This represents our initial belief about the probability of a hypothesis $H$ before we see any new data or evidence. It reflects what we know about the situation prior to the latest information.

2. **Likelihood $P(E|H)$**: This is the probability of observing the evidence $E$ given that the hypothesis $H$ is true. It measures how likely we are to observe the evidence if the hypothesis holds.

3. **Evidence $P(E)$**: Also known as the marginal likelihood, this is the probability of observing the evidence under all possible hypotheses. It can be calculated by considering all possible states of the world (all hypotheses) and summing or integrating over them.

4. **Posterior Probability $P(H|E)$**: This is what Bayesian inference aims to compute - the probability of the hypothesis $H$ after taking into account the new evidence $E$ and our prior belief. This represents our updated belief about the hypothesis in light of the new evidence.

Bayes' theorem provides the mathematical formula for computing the posterior probability:

$ P(H|E) = \frac{P(E|H) \cdot P(H)}{P(E)} $



${\displaystyle {\text{posterior}}={\frac {{\text{prior}}\times {\text{likelihood}}}{\text{evidence}}}\,}$

### 2. Naive Bayes Classifier
A Naive Bayes classifier is a probabilistic machine learning model that's used for classification tasks. It is based on Bayes' theorem and the assumption of independence among features.
The Naive Bayes classifier applies Bayes' theorem with the "naive" assumption that all features are independent given the class. Despite this often unrealistic assumption, Naive Bayes classifiers perform surprisingly well in many real-world applications.

### 3. Calculation Steps

#### Training Phase:
1. **Calculate Prior Probabilities:**
   $ P(C_k) = \frac{\text{Number of instances in class } C_k}{\text{Total number of instances}} $
   This is the probability of each class in the dataset.

2. **Calculate Likelihood:**
   For each feature, compute the likelihood of that feature given each class. If the feature is continuous, you might assume a Gaussian distribution to compute the likelihood:
   $ P(x_i|C_k) = \frac{1}{\sqrt{2\pi\sigma_k^2}} \exp\left(-\frac{(x_i - \mu_k)^2}{2\sigma_k^2}\right) $
   Where $ \mu_k $ and $ \sigma_k $ are the mean and standard deviation of the feature $ x_i $ for class $ C_k $.

3. **Estimate Conditional Probabilities:**
   For categorical features, you simply count the occurrences:
   $ P(x_i|C_k) = \frac{\text{Number of instances where } x_i \text{ occurs in class } C_k}{\text{Number of instances in class } C_k} $

#### Prediction Phase:
1. **Calculate Posterior Probability:**
   For a new instance $ x = (x_1, x_2, ..., x_n) $, compute the posterior probability for each class:
   $ P(C_k|x) \propto P(C_k) \prod_{i=1}^{n} P(x_i|C_k) $
   This involves multiplying the prior probability of the class by the likelihoods of each feature given that class.

2. **Choose the Class with the Highest Posterior Probability:**
   $ \hat{C} = \arg\max_{C_k} P(C_k|x) $

### 4. Types of Naive Bayes Classifiers

#### Gaussian Naive Bayes:
- Used when features are continuous and assumed to follow a Gaussian distribution.

#### Multinomial Naive Bayes:
- Suitable for discrete count data, often used in text classification where features are word frequencies.

#### Bernoulli Naive Bayes:
- Used for binary/boolean features, where features are binary variables.

### 5. Advantages and Disadvantages

#### Advantages:
- **Simple and Fast:** Easy to implement and computationally efficient.
- **Works well with high-dimensional data:** Performs well in text classification and other applications with many features.
- **Robust to irrelevant features:** The independence assumption helps reduce the impact of irrelevant features.

#### Disadvantages:
- **Independence Assumption:** The strong independence assumption is rarely true in real-world data, which can affect performance.
- **Zero Probability Problem:** If a feature value was not seen in the training set for a particular class, it might lead to a zero probability in the product. This is typically handled by techniques like Laplace smoothing.

In summary, the Naive Bayes classifier is a simple yet powerful classification algorithm that leverages Bayes' theorem with an independence assumption. It is widely used in various domains, especially in text classification and spam filtering, due to its effectiveness and efficiency.

### Example: Fruit classification
We'll use a small dataset for classifying whether a fruit is an "Apple" or an "Orange" based on its features: weight (heavy or light) and texture (smooth or rough).

### Dataset
| Fruit  | Weight | Texture | Class   |
|--------|--------|---------|---------|
| Fruit1 | Heavy  | Smooth  | Apple   |
| Fruit2 | Heavy  | Rough   | Orange  |
| Fruit3 | Light  | Smooth  | Apple   |
| Fruit4 | Light  | Rough   | Orange  |
| Fruit5 | Heavy  | Smooth  | Apple   |
| Fruit6 | Heavy  | Rough   | Orange  |

### Step-by-Step Calculation

#### Step 1: Calculate Prior Probabilities
The prior probability of each class is calculated as the proportion of each class in the dataset.

- $ P(Apple) = \frac{\text{Number of Apples}}{\text{Total number of fruits}} = \frac{3}{6} = 0.5 $
- $ P(Orange) = \frac{\text{Number of Oranges}}{\text{Total number of fruits}} = \frac{3}{6} = 0.5 $

#### Step 2: Calculate Likelihoods
We calculate the likelihood of each feature given each class. 

- For Weight:
  - $ P(\text{Heavy}|\text{Apple}) = \frac{\text{Number of Heavy Apples}}{\text{Total number of Apples}} = \frac{2}{3} \approx 0.67 $
  - $ P(\text{Light}|\text{Apple}) = \frac{\text{Number of Light Apples}}{\text{Total number of Apples}} = \frac{1}{3} \approx 0.33 $
  - $ P(\text{Heavy}|\text{Orange}) = \frac{\text{Number of Heavy Oranges}}{\text{Total number of Oranges}} = \frac{2}{3} \approx 0.67 $
  - $ P(\text{Light}|\text{Orange}) = \frac{\text{Number of Light Oranges}}{\text{Total number of Oranges}} = \frac{1}{3} \approx 0.33 $

- For Texture:
  - $ P(\text{Smooth}|\text{Apple}) = \frac{\text{Number of Smooth Apples}}{\text{Total number of Apples}} = \frac{2}{3} \approx 0.67 $
  - $ P(\text{Rough}|\text{Apple}) = \frac{\text{Number of Rough Apples}}{\text{Total number of Apples}} = \frac{1}{3} \approx 0.33 $
  - $ P(\text{Smooth}|\text{Orange}) = \frac{\text{Number of Smooth Oranges}}{\text{Total number of Oranges}} = \frac{1}{3} \approx 0.33 $
  - $ P(\text{Rough}|\text{Orange}) = \frac{\text{Number of Rough Oranges}}{\text{Total number of Oranges}} = \frac{2}{3} \approx 0.67 $

#### Step 3: Classify a New Fruit
Let's classify a new fruit with the features: Weight = Heavy and Texture = Rough.

1. **Calculate the posterior probability for Apple:**
   $ P(\text{Apple}|\text{Heavy}, \text{Rough}) \propto P(\text{Apple}) \times P(\text{Heavy}|\text{Apple}) \times P(\text{Rough}|\text{Apple}) $
   $ P(\text{Apple}|\text{Heavy}, \text{Rough}) \propto 0.5 \times 0.67 \times 0.33 \approx 0.1106 $

2. **Calculate the posterior probability for Orange:**
   $ P(\text{Orange}|\text{Heavy}, \text{Rough}) \propto P(\text{Orange}) \times P(\text{Heavy}|\text{Orange}) \times P(\text{Rough}|\text{Orange}) $
   $ P(\text{Orange}|\text{Heavy}, \text{Rough}) \propto 0.5 \times 0.67 \times 0.67 \approx 0.2245 $

#### Step 4: Choose the Class with the Highest Posterior Probability
Since $ P(\text{Orange}|\text{Heavy}, \text{Rough}) \approx 0.2245 $ is greater than $ P(\text{Apple}|\text{Heavy}, \text{Rough}) \approx 0.1106 $, we classify the new fruit as an **Orange**.

### Summary
- **Prior Probabilities:**
  - $ P(Apple) = 0.5 $
  - $ P(Orange) = 0.5 $
- **Likelihoods:**
  - $ P(\text{Heavy}|\text{Apple}) \approx 0.67 $
  - $ P(\text{Light}|\text{Apple}) \approx 0.33 $
  - $ P(\text{Heavy}|\text{Orange}) \approx 0.67 $
  - $ P(\text{Light}|\text{Orange}) \approx 0.33 $
  - $ P(\text{Smooth}|\text{Apple}) \approx 0.67 $
  - $ P(\text{Rough}|\text{Apple}) \approx 0.33 $
  - $ P(\text{Smooth}|\text{Orange}) \approx 0.33 $
  - $ P(\text{Rough}|\text{Orange}) \approx 0.67 $
- **Posterior Probabilities:**
  - $ P(\text{Apple}|\text{Heavy}, \text{Rough}) \approx 0.1106 $
  - $ P(\text{Orange}|\text{Heavy}, \text{Rough}) \approx 0.2245 $

The new fruit is classified as an **Orange**.

### Example: Person classification

1. **Gaussian Probability Density Function (PDF):**
   $
   P(x|\mu,\sigma) = \frac{1}{\sqrt{2\pi\sigma^2}} e^{-\frac{(x-\mu)^2}{2\sigma^2}}
   $
   where $ \mu $ is the mean, $ \sigma $ is the standard deviation, and $ x $ is the variable.

2. **Posterior Probability (Bayes' Theorem):**
   $
   P(class|data) = \frac{P(class) \cdot P(data|class)}{P(data)}
   $
   where $ P(class) $ is the prior probability of the class, $ P(data|class) $ is the likelihood, and $ P(data) $ is the evidence.

3. **Evidence:**
   $
   P(data) = P(male) \cdot P(height|male) \cdot P(weight|male) \cdot P(foot\_size|male) + P(female) \cdot P(height|female) \cdot P(weight|female) \cdot P(foot\_size|female)
   $

### Documentation
- **Function `gaussian_pdf_function(mu, sigma, x)`**: Computes the Gaussian PDF for a given mean (`mu`), standard deviation (`sigma`), and value (`x`).
- **Training Data**: Lists `height_male`, `weight_male`, `foot_size_male`, `height_female`, `weight_female`, and `foot_size_female` contain the training data for males and females.
- **Mean and Variance**: Calculations of mean and variance for the height, weight, and foot size of males and females.
- **Plotting**: Gaussian PDF curves for the height, weight, and foot size of males and females.
- **Testing**: Computes the posterior probabilities for a test case with

 height 6, weight 130, and foot size 8, using the Naive Bayes classifier to determine the class (male or female) based on the maximum posterior probability.

```python
import numpy as np
import matplotlib.pyplot as plt

#-------------------------------------------------------------------------------------------------------------------------------
# Example taken from https://en.wikipedia.org/wiki/Naive_Bayes_classifier#Examples
# Data:
# Sex     Height(feet)    Weight(lbs)    Foot Size(inches)
# male      6                180                12
# male      5.92 (5'11")     190                11
# male      5.58 (5'7")      170                12
# male      5.92 (5'11")     165                10
# female    5                100                6
# female    5.5 (5'6")       150                8
# female    5.42 (5'5")      130                7
# female    5.75 (5'9")      150                9

def gaussian_pdf_function(mu, sigma, x):
    """
    Calculate the Gaussian probability density function.
    
    Parameters:
    mu (float): Mean of the distribution
    sigma (float): Standard deviation of the distribution
    x (float): Value at which to calculate the PDF

    Returns:
    float: Value of the Gaussian PDF at x
    """
    pi = np.pi
    e = np.e
    denominator = 1 / ((2 * pi * sigma ** 2) ** 0.5)
    exponent = -(x - mu) ** 2 / (2 * sigma ** 2)
    y = denominator * e ** exponent
    return y

# Training Data for males
height_male = [6, 5.92, 5.58, 5.92]
weight_male = [180, 190, 170, 165]
foot_size_male = [12, 11, 12, 10]

# Training Data for females
height_female = [5, 5.5, 5.42, 5.75]
weight_female = [100, 150, 130, 150]
foot_size_female = [6, 8, 7, 9]

# Mean and Variance for males
height_male_mean = np.mean(height_male)
weight_male_mean = np.mean(weight_male)
foot_size_male_mean = np.mean(foot_size_male)

height_male_variance = np.var(height_male)
weight_male_variance = np.var(weight_male)
foot_size_male_variance = np.var(foot_size_male)

# Mean and Variance for females
height_female_mean = np.mean(height_female)
weight_female_mean = np.mean(weight_female)
foot_size_female_mean = np.mean(foot_size_female)

height_female_variance = np.var(height_female)
weight_female_variance = np.var(weight_female)
foot_size_female_variance = np.var(foot_size_female)

# Plotting Gaussian PDFs for male data
fig = plt.figure()

# Height for males
x_min_range = 4
x_max_range = 7
step = 0.05
x = np.arange(x_min_range, x_max_range, step)
y = gaussian_pdf_function(height_male_mean, height_male_variance ** 0.5, x)
ax1 = fig.add_subplot(231)
ax1.set_ylabel('Height (male)')
ax1.plot(x, y, 'r-')

# Weight for males
x_min_range = 40
x_max_range = 220
step = 1
x = np.arange(x_min_range, x_max_range, step)
y = gaussian_pdf_function(weight_male_mean, weight_male_variance ** 0.5, x)
ax2 = fig.add_subplot(232)
ax2.set_ylabel('Weight (male)')
ax2.plot(x, y, 'k-')

# Foot size for males
x_min_range = 4
x_max_range = 20
step = 0.1
x = np.arange(x_min_range, x_max_range, step)
y = gaussian_pdf_function(foot_size_male_mean, foot_size_male_variance ** 0.5, x)
ax3 = fig.add_subplot(233)
ax3.set_ylabel('Foot size (male)')
ax3.plot(x, y, 'b-')

# Plotting Gaussian PDFs for female data
# Height for females
x_min_range = 4
x_max_range = 7
step = 0.05
x = np.arange(x_min_range, x_max_range, step)
y = gaussian_pdf_function(height_female_mean, height_female_variance ** 0.5, x)
ax4 = fig.add_subplot(234)
ax4.set_ylabel('Height (female)')
ax4.plot(x, y, 'k-')

# Weight for females
x_min_range = 40
x_max_range = 220
step = 1
x = np.arange(x_min_range, x_max_range, step)
y = gaussian_pdf_function(weight_female_mean, weight_female_variance ** 0.5, x)
ax5 = fig.add_subplot(235)
ax5.set_ylabel('Weight (female)')
ax5.plot(x, y, 'k-')

# Foot size for females
x_min_range = 4
x_max_range = 20
step = 0.1
x = np.arange(x_min_range, x_max_range, step)
y = gaussian_pdf_function(foot_size_female_mean, foot_size_female_variance ** 0.5, x)
ax6 = fig.add_subplot(236)
ax6.set_ylabel('Foot size (female)')
ax6.plot(x, y, 'b-')

plt.show()

# Testing Data
height = 6
weight = 130
foot_size = 8

# Prior probabilities
P_male = 0.5
P_female = 0.5

# Likelihoods
# P(height|male)
P_height_male = gaussian_pdf_function(height_male_mean, height_male_variance ** 0.5, height)

# P(weight|male)
P_weight_male = gaussian_pdf_function(weight_male_mean, weight_male_variance ** 0.5, weight)

# P(foot_size|male)
P_foot_size_male = gaussian_pdf_function(foot_size_male_mean, foot_size_male_variance ** 0.5, foot_size)

# P(height|female)
P_height_female = gaussian_pdf_function(height_female_mean, height_female_variance ** 0.5, height)

# P(weight|female)
P_weight_female = gaussian_pdf_function(weight_female_mean, weight_female_variance ** 0.5, weight)

# P(foot_size|female)
P_foot_size_female = gaussian_pdf_function(foot_size_female_mean, foot_size_female_variance ** 0.5, foot_size)

# Evidence
evidence = (P_male * P_height_male * P_weight_male * P_foot_size_male + 
            P_female * P_height_female * P_weight_female * P_foot_size_female)

# Posterior probabilities
male_posterior = (P_male * P_height_male * P_weight_male * P_foot_size_male) / evidence
female_posterior = (P_female * P_height_female * P_weight_female * P_foot_size_female) / evidence

print(f"Male posterior is: {male_posterior}, Female posterior is: {female_posterior}")
print("Based on the posterior probabilities, the data is classified as female.")
```

```python
#Solving problem using sklearn package
import numpy as np
from sklearn.naive_bayes import GaussianNB

# Training Data, male
height_male = [6, 5.92, 5.58, 5.92]
weight_male = [180, 190, 170, 165]
foot_size_male = [12, 11, 12, 10]

# Training Data, female
height_female = [5, 5.5, 5.42, 5.75]
weight_female = [100, 150, 130, 150]
foot_size_female = [6, 8, 7, 9]

# Combining the data into a single array
height = np.hstack((height_male, height_female))
weight = np.hstack((weight_male, weight_female))
foot_size = np.hstack((foot_size_male, foot_size_female))
record = np.column_stack((height, weight, foot_size))

class_labels = np.array([1, 1, 1, 1, 2, 2, 2, 2])

test_data = np.array([6, 130, 8])
test_data = test_data.reshape(1, -1)

# Training the model
clf = GaussianNB()
clf.fit(record, class_labels)

# Predicting the class of the test data
predicted_class = clf.predict(test_data)

print(f"Then our data must belong to the class number: {predicted_class[0]}")
```
