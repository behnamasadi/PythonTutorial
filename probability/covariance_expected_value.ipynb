## Expected Value

The expected value represents the long-term average or mean value of a random variable over many trials or observations. It is essentially a weighted average, where each possible outcome of the random variable is weighted by its probability of occurring.

For a discrete random variable $X$ with possible values $x_1, x_2, \ldots, x_n$ and corresponding probabilities $P(X = x_1), P(X = x_2), \ldots, P(X = x_n)$, the expected value $E(X)$ is given by:

$ E(X) = \sum_{i=1}^{n} x_i P(X = x_i) $

For a continuous random variable $X$ with probability density function $f(x)$, the expected value $E(X)$ is given by:

$ E(X) = \int_{-\infty}^{\infty} x f(x) \, dx $

### Important Properties of Expected Value

1. **Linearity of Expectation:**
   The expected value operator is linear, meaning that for any random variables $X$ and $Y$, and any constants $a$ and $b$:

   $ E(aX + bY) = aE(X) + bE(Y) $

   This property also extends to the sum of multiple random variables:

   $ E\left(\sum_{i=1}^{n} a_i X_i\right) = \sum_{i=1}^{n} a_i E(X_i) $

2. **Expectation of a Constant:**
   If $c$ is a constant, then the expected value of $c$ is just $c$:

   $ E(c) = c $

3. **Expected Value of the Sum of Random Variables:**
   For any random variables $X$ and $Y$:

   $ E(X + Y) = E(X) + E(Y) $

   This holds true whether $X$ and $Y$ are independent or not.

4. **Expected Value of a Product of Independent Random Variables:**
   If $X$ and $Y$ are independent random variables, then:

   $ E(XY) = E(X) \cdot E(Y) $

5. **Non-Negativity:**
   If $X$ is a non-negative random variable (i.e., $X \geq 0$ always), then:

   $ E(X) \geq 0 $

### Example Calculation

Consider a discrete random variable $X$ that can take values 1, 2, and 3 with probabilities 0.2, 0.5, and 0.3, respectively. The expected value $E(X)$ can be calculated as follows:

$ E(X) = 1 \cdot 0.2 + 2 \cdot 0.5 + 3 \cdot 0.3 $
$ E(X) = 0.2 + 1 + 0.9 $
$ E(X) = 2.1 $

So, the expected value $E(X)$ is 2.1.

# Variance

Variance describes the spread or dispersion of a set of data points. It quantifies how much the values in a dataset deviate from the mean (average) of the dataset. A higher variance indicates that the data points are more spread out from the mean, while a lower variance indicates that they are closer to the mean.

### Population Variance

The variance of a population (denoted as $\sigma^2$) is calculated using the following steps:

1. Find the mean (average) of the population.
2. Subtract the mean from each data point to find the deviation of each data point from the mean.
3. Square each deviation to eliminate negative values.
4. Find the average of these squared deviations.

The formula for the population variance is:

$ \sigma^2 = \frac{1}{N} \sum_{i=1}^{N} (x_i - \mu)^2 $

Where:
- $ \sigma^2 $ is the population variance.
- $ N $ is the number of data points in the population.
- $ x_i $ represents each individual data point.
- $ \mu $ is the mean of the population.
- $\sum$ denotes the summation.


Since we can express expected value $=\mathbb{E}[X]$ as function: $f(X) = \frac{1}{N} \sum_{i=1}^{N} (x_i - \mu) $, we can write $ \sigma^2 = \frac{1}{N} \sum_{i=1}^{N} (x_i - \mu)^2 $ for a random variable $X$:

$ \text{Var}(X) = \mathbb{E}[(X - \mathbb{E}[X])^2] $

Hereâ€™s the breakdown of the formula:

1. $\mathbb{E}[X]$ is the expected value (mean) of the random variable $X$.
2. $X - \mathbb{E}[X]$ represents the deviation of $X$ from its mean.
3. $(X - \mathbb{E}[X])^2$ is the square of the deviation, which ensures that the measure is non-negative and emphasizes larger deviations more.
4. $\mathbb{E}[(X - \mathbb{E}[X])^2]$ is the expected value of these squared deviations, providing an average measure of dispersion around the mean.

An alternative expression for variance using the properties of expected value is:

$ \text{Var}(X) = \mathbb{E}[X^2] - (\mathbb{E}[X])^2 $

In this form:

1. $\mathbb{E}[X^2]$ is the expected value of the square of the random variable $X$.
2. $(\mathbb{E}[X])^2$ is the square of the expected value of $X$.

## Covariance
Covariance is a measure of the joint variability of two random variables

${\displaystyle {\begin{aligned}\operatorname {cov} (X,Y)&=\operatorname {E} \left[\left(X-\operatorname {E} \left[X\right]\right)\left(Y-\operatorname {E} \left[Y\right]\right)\right]\\&=\operatorname {E} \left[XY-X\operatorname {E} \left[Y\right]-\operatorname {E} \left[X\right]Y+\operatorname {E} \left[X\right]\operatorname {E} \left[Y\right]\right]\\&=\operatorname {E} \left[XY\right]-\operatorname {E} \left[X\right]\operatorname {E} \left[Y\right]-\operatorname {E} \left[X\right]\operatorname {E} \left[Y\right]+\operatorname {E} \left[X\right]\operatorname {E} \left[Y\right]\\&=\operatorname {E} \left[XY\right]-\operatorname {E} \left[X\right]\operatorname {E} \left[Y\right],\end{aligned}}}$

To express the covariance matrix in the form $[X - \mu_X][X - \mu_X]^T$ involves a matrix calculation where $X$ is the matrix of observations and $\mu_X$ is the matrix of mean values repeated for each observation.

### Given Data and Definitions

- $X$ (Data matrix where each row is a variable, and each column is an observation):
  $
  X = \begin{bmatrix}
  4.0 & 4.2 & 3.9 & 4.3 & 4.1 \\
  2.0 & 2.1 & 2.0 & 2.1 & 2.2 \\
  0.6 & 0.59 & 0.58 & 0.62 & 0.63
  \end{bmatrix}
  $

- $\mu_X$ (Mean vector, each mean repeated across corresponding rows):
  $
  \mu_X = \begin{bmatrix}
  4.1 & 4.1 & 4.1 & 4.1 & 4.1 \\
  2.08 & 2.08 & 2.08 & 2.08 & 2.08 \\
  0.604 & 0.604 & 0.604 & 0.604 & 0.604
  \end{bmatrix}
  $

### Calculation

1. **Center the data matrix $X$ by subtracting $\mu_X$:**
   $
   X - \mu_X = \begin{bmatrix}
   4.0 - 4.1 & 4.2 - 4.1 & 3.9 - 4.1 & 4.3 - 4.1 & 4.1 - 4.1 \\
   2.0 - 2.08 & 2.1 - 2.08 & 2.0 - 2.08 & 2.1 - 2.08 & 2.2 - 2.08 \\
   0.6 - 0.604 & 0.59 - 0.604 & 0.58 - 0.604 & 0.62 - 0.604 & 0.63 - 0.604
   \end{bmatrix}
   $

2. **Compute the product $[X - \mu_X][X - \mu_X]^T$:**

   $ [X - \mu_X][X - \mu_X]^T = \frac{1}{4} (X - \mu_X)(X - \mu_X)^T  $
   The factor of $\frac{1}{4}$ is used because the standard formula for covariance uses $n-1$ (with $n$ being the number of observations) as the denominator to get an unbiased estimate.

Let's calculate this matrix product to verify that it yields the covariance matrix:

Centered Data Matrix $(X - \mu_X)$: 

The matrix $X - \mu_X$ where each entry is the deviation of each observation from the mean is given by:
$
X - \mu_X = \begin{bmatrix}
-0.1 & 0.1 & -0.2 & 0.2 & 0.0 \\
-0.08 & 0.02 & -0.08 & 0.02 & 0.12 \\
-0.004 & -0.014 & -0.024 & 0.016 & 0.026
\end{bmatrix}
$

Covariance Matrix Calculation $[X - \mu_X][X - \mu_X]^T$: 

The covariance matrix calculated using the formula $\frac{1}{n-1} (X - \mu_X)(X - \mu_X)^T$ matches the previous calculation:
$
\begin{bmatrix}
0.025 & 0.0075 & 0.00175 \\
0.0075 & 0.007 & 0.00135 \\
0.00175 & 0.00135 & 0.00043
\end{bmatrix}
$

This matrix is exactly the covariance matrix we obtained earlier, verifying that both approaches are consistent. The diagonal entries represent the variances of each variable, and the off-diagonal entries represent the covariances between different variables.

### Interpretation

Each element $ \text{Cov}(X_i, X_j) $ of the covariance matrix represents the covariance between variables $ X_i $ and $ X_j $. The covariance provides insights into how much the variables change together. A positive covariance indicates that the variables tend to move in the same direction, whereas a negative covariance suggests they move in opposite directions.

- The diagonal entries (variances) tell how spread out the data is around the mean for each variable.
- The off-diagonal entries (covariances) tell how much two variables change together, providing a sense of their linear relationship. 

This matrix is key in many statistical analyses, including principal component analysis (PCA) and multivariate analysis of variance (MANOVA).

## Different Covariance Structures

In the Gaussian Mixture Model (GMM), the covariance parameter determines the shape, volume, and orientation of the component Gaussians in the mixture. There are three main types of covariance matrices that can be used in GMMs, each imposing different constraints on the component Gaussians:

1. **Spherical**: In this case, the covariance matrix is proportional to the identity matrix. This means all directions have the same variance, and the resulting component is a sphere in the feature space. The assumption here is that the data points in each cluster are spread out uniformly in all directions, with the same variance. This type is computationally the least expensive but also the least flexible in terms of fitting complex data structures.


**Spherical Covariance**:
$ \Sigma = \sigma^2 I $


2. **Diagonal**: The covariance matrices are diagonal, meaning that the features are assumed to be uncorrelated, but each feature can have its own variance. This allows each component Gaussian to have its own variance along each dimension but constrains the shape of the component to be aligned with the axes of the feature space. It's more flexible than the spherical covariance model since it can model elongated clusters but still assumes that the principal axes of the data align with the feature axes.

**Diagonal Covariance**:
$ \Sigma = \text{diag}(\sigma_1^2, \sigma_2^2, \ldots, \sigma_d^2) $


3. **Full (or General)**: The full covariance matrix allows each component to have a distinct covariance matrix, meaning that the variances and covariances can all be different. This is the most flexible model, allowing for ellipsoidal components that can be oriented in any direction in the feature space. However, it is also the most computationally expensive and requires the most data to estimate accurately, as the number of parameters grows quadratically with the number of features.


**Full Covariance**:
$ \Sigma = \begin{pmatrix} \sigma_{11}^2 & \sigma_{12} & \cdots & \sigma_{1d} \\ \sigma_{21} & \sigma_{22}^2 & \cdots & \sigma_{2d} \\ \vdots & \vdots & \ddots & \vdots \\ \sigma_{d1} & \sigma_{d2} & \cdots & \sigma_{dd}^2 \end{pmatrix} $


4. **Tied**: All components share the same general covariance matrix. This means every component Gaussian has the same shape, volume, and orientation, but they can be centered at different points in the feature space. This model strikes a balance between the computational complexity of the full covariance model and the simplicity of spherical or diagonal covariances, by imposing the same covariance structure across all mixture components.

Choosing the right type of covariance structure depends on the complexity of the data, the computational resources available, and the specific requirements of the application. Spherical and diagonal covariances are computationally more efficient and may suffice for simpler or higher-dimensional datasets where the number of data points per dimension is low. Full covariance models are more appropriate for capturing complex relationships in the data but require more data and computational power to estimate accurately. The tied model offers a compromise, applying the same shape constraints across components, which can be useful when you want to model complex structures but have limited data.



```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import multivariate_normal

# Define the mean and covariance matrices for each type
mean = np.array([0, 0])
cov_spherical = np.eye(2)  # Identity matrix for spherical
cov_diagonal = np.array([[1, 0], [0, 3]])  # Diagonal covariance
cov_full = np.array([[1, 0.8], [0.8, 1]])  # Full covariance

# Create a grid of points
x, y = np.mgrid[-3:3:.01, -3:3:.01]
pos = np.dstack((x, y))

# Define the Gaussian distributions
pdf_spherical = multivariate_normal(mean, cov_spherical).pdf(pos)
pdf_diagonal = multivariate_normal(mean, cov_diagonal).pdf(pos)
pdf_full = multivariate_normal(mean, cov_full).pdf(pos)

# Plotting
fig, ax = plt.subplots(1, 3, figsize=(18, 6))

ax[0].contourf(x, y, pdf_spherical)
ax[0].set_title('Spherical Covariance')

ax[1].contourf(x, y, pdf_diagonal)
ax[1].set_title('Diagonal Covariance')

ax[2].contourf(x, y, pdf_full)
ax[2].set_title('Full Covariance')

plt.show()
```
