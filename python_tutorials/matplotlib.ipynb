## **1. Figure: The Top-Level Container**

The `figure` is the outermost container.  Think of it as the canvas on which you'll draw your plots.  You create a figure using `plt.figure()`.  It can hold one or more `Axes` (which are the actual plotting areas).

Create a new figure, or activate an existing figure:
- num: int or str, optional A unique identifier for the figure.
- figsize sets width and height in inches. 

```python
import matplotlib.pyplot as plt
plt.figure(num=None, figsize(float, float))
```
If a figure with that identifier already exists, this figure is made active and returned. An integer refers to the  `Figure.number` attribute, a string refers to the figure label.





## **2. Axes: The Plotting Area**

`Axes` (not to be confused with `axis`) are the individual plotting areas within a figure. It’s a part of the Figure, nothing but a subplot. An Axes object represents a single plot. It's the region where your data is actually plotted.   Think of it as the canvas within a figure where you draw your graphs.  Each `Axes` object is where you'll plot your data (lines, scatter plots, bars, etc.).  A figure can contain multiple `Axes` (subplots).

To create axes in the figure, call `subplots()`:
- `subplots()`:  `plt.subplots(nrows, ncols, index)` is a common way to create a grid of subplots.  `nrows` and `ncols` specify the grid dimensions, and `index` specifies the current subplot (starting from 1).

```python
    fig, axes = plt.subplots(2, 1, figsize=(8, 6)) # axes will be an array of Axes objects
    ax1, ax2 = axes #unpacking the Axes
```

- `add_subplot()`: create the figure first and then call `add_subplot()`:


```python
fig=plt.figure(figsize=[4,6])
ax1 = fig.add_subplot(2, 1, 1)  # Creates the first subplot (top)
ax2 = fig.add_subplot(2, 1, 2)  # Creates the second subplot (bottom)

```



## **3. Axis (and Ticks): The Data Scales**

*   An `Axes` object has two (or three for 3D plots) `Axis` objects: the x-axis and the y-axis (and the z-axis).  The `Axis` objects control the data limits and ticks.

*   `xlim()` and `ylim()`: `ax.set_xlim(xmin, xmax)` and `ax.set_ylim(ymin, ymax)` set the limits of the x and y axes, respectively.  You can also use `plt.xlim()` and `plt.ylim()` if you want to affect the current axes.

    ```python
    ax1.set_xlim(0, 10)
    ax1.set_ylim(-1, 1)
    ```

*   `ticks`: Ticks are the markers on the axes that indicate specific data values. You can control their location and labels.
    ```python
    ax1.set_xticks([0, 2, 4, 6, 8, 10]) # Set where ticks appear
    ax1.set_xticklabels(['0', '2', '4', '6', '8', '10']) # set what the ticks are labelled as
    ```

**4. Titles and Labels**

*   `set_title()`: `ax.set_title("Plot Title")` sets the title for the individual `Axes` (subplot).

*   `set_xlabel()` and `set_ylabel()`: `ax.set_xlabel("X-axis Label")` and `ax.set_ylabel("Y-axis Label")` set the labels for the x and y axes.

    ```python
    ax1.set_title("Sine Wave")
    ax1.set_xlabel("Time")
    ax1.set_ylabel("Amplitude")
    ```

**5. Legend**

*   `legend()`: `ax.legend()` adds a legend to the `Axes`.  You need to label your plot elements (e.g., lines) using the `label` keyword argument when plotting.

    ```python
    line, = ax1.plot([1, 2, 3], [4, 5, 6], label="Data") # the comma after line is important!
    ax1.legend()
    ```

**6. Grid**

*   `grid()`: `ax.grid(True)` turns the grid lines on for the `Axes`.



## Examples
### figure()

````python
import matplotlib.pyplot as plt
import numpy as np

fig1 = plt.figure(figsize=[4, 6])
fig2 = plt.figure(figsize=[5, 7])

plt.figure(1)
print ("current figure = " + str(plt.gcf().number))

plt.figure(2)
print ("current figure = " + str(plt.gcf().number))

# now switch back to figure 1 and make some changes
x=np.linspace(start=0, stop=10, num=50)
plt.figure(1)
plt.plot(x,x**2)

# now switch back to figure 2 and make some changes
plt.figure(2)
plt.plot(x,x**3)
# Get the current Axes instance on the current figure
ax = plt.gca()
ax.set_title('y=x**3')
plt.show()

print ("Figure numbers are = " + str(plt.get_fignums()))
````

### axis, xlim, ylim limit

````python
import matplotlib.pyplot as plt
import numpy as np


fig = plt.figure()

ax1 = fig.add_subplot(1, 2, 1)
ax2 = fig.add_subplot(1, 2, 2)

x = np.arange(-4*np.pi, 4*np.pi, 0.1)
y = np.sin(x)
y_shifted = np.sin(x-np.pi/4)

xmin = 0
xmax = 4*np.pi
ymin = 0
ymax = 2


ax1.plot(x, y, color='r')
ax1.plot(x, y_shifted, color='b')


ax2.axis([xmin, xmax, ymin, ymax])
ax2.plot(x, y, color='r')
ax2.plot(x, y_shifted, color='b')


plt.show()
````

### suptitle, legend, label, set_title, set_xlabel, ticks, grid


- legend adds the labels you specified when creating plot elements to the visualization `label="this is sin"`. When you call `plt.legend()`, Matplotlib looks for all the plot elements that have associated labels.
- Common locations include 'upper right', 'upper left', 'lower right', 'lower left', 'best' (Matplotlib automatically chooses the best location), and many more.

````python
import matplotlib.pyplot as plt
import numpy as np


x = np.arange(0.0, 5.0, 0.1)

fig, ax = plt.subplots(nrows=1, ncols=1)
fig.suptitle('fig title', fontsize=10)
ax.plot(x, np.sin(x), label="this is sin")
ax.legend(loc="best")
ax.set_aspect('equal')
ax.set_title('Line Collection with mapped colors')
ax.set_xlabel("this is x")
ax.set_ylabel("this is y")
ax.xaxis.set_ticks_position('top')
ax.yaxis.set_ticks_position('right')
ax.grid(True)
plt.show()
````

## plot_surface()


`numpy.meshgrid` is a function in the NumPy library that's used to create a rectangular grid out of one-dimensional arrays. Imagine you have two 1D arrays, `x` and `y`. meshgrid creates two 2D arrays, `X` and `Y`, such that:

- X contains copies of x along the rows, and
- Y contains copies of y along the columns.

Essentially, it creates **all possible combinations of points from the input arrays**.


````python
x = np.array([1, 2, 3])
y = np.array([4, 5, 6, 7])

X, Y = np.meshgrid(x, y)

print("X:\n", X)
print("Y:\n", Y)
````

````python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import colormaps

print(colormaps)

X = np.arange(-10, 10, 0.25)
Y = np.arange(-10, 10, 0.25)
X, Y = np.meshgrid(X, Y)
Z = np.sin(np.sqrt((X-1)**2 + (Y-3)**2))

fig = plt.figure()
ax = fig.add_subplot(projection='3d')
surf = ax.plot_surface(X, Y, Z, cmap='twilight')


plt.show()
````

### quiver()
Quiver plots in Matplotlib are specifically used to visualize vector fields. This means they are ideal for representing data that has both a magnitude and a direction. For instance, analyzing stress gradients in materials.



````python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np


# [x_start, y_start, z_start, x_length, y_length, z_length]
soa = np.array([[0, 0, 1, 1, -2, 0], [0, 0, 2, 1, 1, 0],
                [0, 0, 3, 2, 1, 0], [0, 0, 4, 0.5, 0.7, 0]])\



X, Y, Z, U, V, W = zip(*soa)


fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.quiver(X, Y, Z, U, V, W,length=.005)


ax.autoscale(enable=False, axis='both', tight=None)

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

plt.show()
````
