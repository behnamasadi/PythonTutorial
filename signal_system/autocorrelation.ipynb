# Autocorrelation
Autocorrelation, often denoted by $ R(\tau) $ or $ R_{xx}(\tau) $, is a mathematical tool used to analyze the degree to which a signal correlates with a time-delayed version of itself. It can provide insights into repetitive patterns, periodicities, or the presence of certain frequencies in a signal.

Here's a breakdown:

### Definition:

For a continuous-time signal $ x(t) $, the autocorrelation $ R_{xx}(\tau) $ is given by:

$ R_{xx}(\tau) = \int_{-\infty}^{\infty} x(t) x(t+\tau) \, dt $

For a discrete-time signal $ x[n] $, the autocorrelation $ R_{xx}[m] $ is:

$ R_{xx}[m] = \sum_{n=-\infty}^{\infty} x[n] x[n+m] $

Where:
- $ \tau $ is the time delay for the continuous-time signal.
- $ m $ is the sample delay for the discrete-time signal.

### Interpretation:

1. **Zero Delay**: At $ \tau = 0 $ or $ m = 0 $, the autocorrelation gives the energy of the signal. In this case, you're comparing the signal with itself without any delay.

2. **Periodic Signals**: If a signal is periodic, its autocorrelation will also be periodic. Peaks in the autocorrelation function will occur at lags corresponding to the period of the signal. 

3. **Random Signals**: For random signals like white noise, the autocorrelation will be an impulse at zero lag (indicating the signal's variance) and close to zero otherwise, which indicates that the values are uncorrelated at different time instances.

4. **Stationary vs. Non-stationary Signals**: The autocorrelation function for stationary signals (signals whose statistical properties do not change over time) depends only on the time difference $ \tau $ or $ m $. For non-stationary signals, the autocorrelation function might change with time.

### Applications:

1. **Signal Processing**: It's used to identify periodic components in a signal or to determine the main frequency components.

2. **Communication**: In communication systems, autocorrelation can be used to detect specific signals in a noisy environment.

3. **Economics and Finance**: In time series analysis, autocorrelation can indicate if today's values might be influenced by previous days' values.

4. **Image Processing**: It can be used for pattern recognition and other analysis.

### Example:



### 1. Zero Delay:

For any signal, at zero delay, the autocorrelation is simply the signal's energy.




```python

import numpy as np
import matplotlib.pyplot as plt

def autocorr(x):
    result = np.correlate(x, x, mode='full')
    return result[result.size // 2:]


# Signal
x = np.array([1, 2, 3, 4, 5])
ac = autocorr(x)

plt.stem(ac)
plt.title('Autocorrelation for a Simple Signal')
plt.xlabel('Lag')
plt.ylabel('Autocorrelation')
plt.show()
```

In the plot, the value at **lag 0** is the highest, representing the energy of the signal.


### 2. Periodic Signals:
Let's consider a sinusoidal signal.


```python
t = np.linspace(0, 2*np.pi, 1000, endpoint=False)
x = np.sin(2 * np.pi * 7 * t)  # A sinusoid with frequency 7 Hz
ac = autocorr(x)

plt.stem(ac[:100])  # Showing just the first 100 lags for clarity
plt.title('Autocorrelation for a Sinusoidal Signal')
plt.xlabel('Lag')
plt.ylabel('Autocorrelation')
plt.show()


```

For the sinusoidal signal, you'll see peaks in the autocorrelation at lags corresponding to the period of the sinusoid.


### 3. Random Signals:
Here's an example with white noise.


```python
x = np.random.randn(1000)  # Gaussian white noise
ac = autocorr(x)

plt.stem(ac[:100])  # Showing just the first 100 lags for clarity
plt.title('Autocorrelation for White Noise')
plt.xlabel('Lag')
plt.ylabel('Autocorrelation')
plt.show()
```

For white noise, the autocorrelation is an impulse at lag 0 and nearly zero for other lags, indicating the noise values are uncorrelated.

### 4. Stationary vs. Non-stationary Signals:

A stationary signal doesn't change its statistical properties over time, while a non-stationary one does.






```python
# Stationary Signal: Sinusoid
x_stationary = np.sin(2 * np.pi * 7 * t)
ac_stationary = autocorr(x_stationary)

# Non-stationary Signal: Sinusoid with increasing amplitude
x_non_stationary = t * np.sin(2 * np.pi * 7 * t)
ac_non_stationary = autocorr(x_non_stationary)

plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.stem(ac_stationary[:100] )
plt.title('Autocorrelation for a Stationary Signal')
plt.xlabel('Lag')
plt.ylabel('Autocorrelation')

plt.subplot(1, 2, 2)
plt.stem(ac_non_stationary[:100])
plt.title('Autocorrelation for a Non-stationary Signal')
plt.xlabel('Lag')
plt.ylabel('Autocorrelation')

plt.tight_layout()
plt.show()

```

For the stationary sinusoid, the autocorrelation retains a periodic pattern. For the non-stationary signal, the autocorrelation pattern reflects the changing nature of the signal.

Note: The visual patterns in the autocorrelation plots might vary slightly based on the specifics of the signals, their lengths, and the sampling rate. The above illustrations are meant to give a general idea.
