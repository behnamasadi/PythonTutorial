# Random Walk

A random walk is a mathematical formalization of a path consisting of a succession of random steps. 

**1. Basic Definition:** 
Imagine you're standing at the origin of an integer number line. Flip a coin. If it's heads, you step to the right (i.e., increase your position by one unit). If it's tails, you step to the left (decrease your position by one unit). Each step you take is random and independent of previous steps. The series of steps you take represents a one-dimensional random walk.

**2. Dimensions:**
- **1D**: Like the example above, you move either left or right on a line.
- **2D**: Imagine you're on a grid, like a chessboard. You can move up, down, left, or right. Each move is random.
- **3D**: This is similar to the 2D scenario but now you can also move up or down, in addition to the other directions, in a 3-dimensional space.

**3. Properties:**
- **Independent Steps**: Each step is independent of the previous ones.
- **Symmetric**: In a simple random walk, each direction has an equal probability. For instance, in a 1D random walk, the chance of moving left is the same as moving right.
- **Non-Symmetric**: The probabilities for each direction are not the same. For example, you might have a 60% chance of moving right and a 40% chance of moving left.

**4. Applications and Extensions:**
- **Stock Prices**: The random walk theory in finance suggests that stock prices follow a random walk and thus cannot be predicted with certainty based on historical price patterns.
- **Brownian Motion**: This is a random walk on a continuous domain, representing the motion of particles suspended in a fluid (gas or liquid).
- **Drunkard's Walk**: A colloquial name for a random walk, envisioning a drunk person staggering randomly.
- **Random Walks on Graphs**: Useful in computer science and network theory, this variation sees the walker move from one node to another connected node at random.

**5. Mathematical Implications:**
- For a simple symmetric random walk, it can be shown that the walker will return to the starting point with a probability of 1 (certainty) in 1D and 2D scenarios. However, in 3D, while the walker will come arbitrarily close to the starting point infinitely often, it's not certain that they will return to the exact starting point.
  
**6. Random Walk vs. Markov Chain:**
A random walk is a specific type of Markov Chain. In a Markov Chain, future states depend only on the current state and not on the sequence of states that preceded it. A random walk satisfies this property.




The choice between a Gaussian distribution and a Uniform distribution for generating steps in a random walk depends on the specific scenario or the properties you want the random walk to exhibit. Here are considerations for both:

1. **Gaussian (Normal) Distribution**:
   - If you use a Gaussian distribution to generate the steps, your random walk will exhibit steps of varying magnitude, which could be reflective of more natural or real-world processes where variations can occur in a continuous manner.
   - The Gaussian distribution is defined by its mean (`mu`) and standard deviation (`sigma`), which you can tune to model different levels of variation in the steps.

```python
def random_walk(n_steps=1000, mu=0, sigma=0.1):
    """Generate a random walk of n_steps using a Gaussian distribution."""
    steps = np.random.normal(mu, sigma, n_steps)
    walk = np.cumsum(steps)
    return walk
```

2. **Uniform Distribution**:
   - If you use a Uniform distribution, the steps in your random walk will be of fixed magnitude but random direction, which might be suitable for simpler or more controlled scenarios.
   - In a simple random walk, each step is typically either +1 or -1, which can be modeled using a Uniform distribution.

```python
def random_walk(n_steps=1000):
    """Generate a random walk of n_steps using a Uniform distribution."""
    steps = np.random.choice([-1, 1], size=n_steps)
    walk = np.cumsum(steps)
    return walk
```

In your provided code, you've chosen a Gaussian distribution with a mean of 0 and a standard deviation of 0.1 to generate the steps. This will yield a random walk where each step size follows a normal distribution, leading to a more varied step size compared to a uniform distribution.

It's important to choose the distribution based on the context and the characteristics you want the random walk to have.

````python
import numpy as np
import matplotlib.pyplot as plt

def random_walk_uniform(n_steps=1000):
    """Generate a random walk of n_steps using a Uniform distribution, either +1 or -1."""
    steps = np.random.choice([-1, 1], size=n_steps)
    walk = np.cumsum(steps)
    return walk

def random_walk_gaussian(n_steps=1000, mu=0, sigma=0.1):
    """Generate a random walk of n_steps using a Gaussian distribution."""
    steps = np.random.normal(mu, sigma, n_steps)
    walk = np.cumsum(steps)
    return walk

# Generate a random walk of 1000 steps for each distribution
walk_uniform = random_walk_uniform(1000)
walk_gaussian = random_walk_gaussian(1000)

# Plot the random walks
plt.figure(figsize=(10, 6))
plt.title("1D Random Walks")
plt.xlabel("Number of Steps")
plt.ylabel("Position")
plt.plot(walk_uniform, label='Uniform Distribution')
plt.plot(walk_gaussian, label='Gaussian Distribution')
plt.legend()
plt.grid(True)
plt.show()

````

# Brownian noise, Brownian motion or Wiener process

Brownian noise, also known as Brownian motion or Wiener process, is named after the Scottish botanist Robert Brown.


1. **Mathematical Model**: Brownian motion can be modeled mathematically as a stochastic process. The Wiener process is the most well-known continuous-time process modeling Brownian motion. It's characterized by:
   - Starting at zero.
   - Having independent increments.
   - The motion over any interval of time has a normal distribution with mean zero.

2. **Brownian Noise in Sound**: When we talk about "Brownian noise" in the context of audio or signal processing, we're referring to a signal that has the statistical properties of Brownian motion. Brownian noise is a type of "color" of noise. Other colors of noise include **white noise**, **pink noise**, etc. Brownian noise has a power density that decreases at a rate of 6 dB per octave with increasing frequency (or -20 dB/decade). This means it has more energy at lower frequencies.

3. **Applications**: 
   - **Sound Synthesis**: Brownian noise is sometimes used in sound synthesis and music, especially for creating natural or atmospheric sounds.
   - **Environmental Simulations**: The randomness of Brownian noise can be used to simulate various natural processes.
   - **Signal Processing**: In some advanced applications, the properties of Brownian noise can be used in filtering or processing signals.

4. **Comparison to White Noise**: White noise is another well-known noise "color". It's characterized by having a constant power density across all frequencies. This means if you were to listen to white noise, every octave would have an equal amount of energy, making it sound very "hissy". In contrast, Brownian noise is dominated by its lower frequencies.

Absolutely! The Wiener process $ W(t) $ is one of the foundational concepts of stochastic processes and plays a central role in the theory of Brownian motion. Here's a description of its properties, and I'll break each of them down:

1. **Initial Condition**: $ W(0) = 0 $
   
   This states that the process starts at 0. In the context of Brownian motion, if you were to observe a particle and note its position at the beginning of your observation as your reference, the deviation from this reference is 0.

2. **Independence of Increments**:
   
   The increments $ W(t + \tau) - W(t) $ are independent for non-overlapping intervals of $ t $ and $ \tau $.

   This means that the change in the process over any time interval is independent of the change in any other non-overlapping time interval. In simpler terms, how the particle moves in one interval doesn't depend on how it moved in any previous interval.

3. **Normally Distributed Increments**:
   
   For any $ t > s $, the increment $ W(t) - W(s) $ is normally distributed with mean 0 and variance $ t - s $, i.e., $ W(t) - W(s) \sim N(0, t - s) $.

   This property says that the changes or increments in the Wiener process over any time interval follow a normal (Gaussian) distribution. The mean 0 indicates that there's no drift in any direction, and the variance $ t - s $ signifies that the variability of these increments grows with the length of the time interval.

4. **Continuous Paths**:

   For every realization (or sample path) of $ W(t) $, $ t \geq 0 $, the path is continuous.

   This implies that even though the Wiener process is random and exhibits unpredictable behavior, it doesn't make jumps; it changes in a continuous manner. When visualized, each path is a continuous curve.

5. **Non-Differentiability**:

   The paths of $ W(t) $ are almost surely nowhere differentiable. 

   This is a more technical property. In simple terms, if you try to "zoom in" on any segment of a path of Brownian motion in an attempt to find a smooth portion that you can differentiate (like you would with regular calculus), you'll be disappointed. No matter how much you zoom in, the path remains as jagged and unpredictable as before.

These properties define the Wiener process, which is a continuous-time stochastic process that serves as the mathematical foundation for Brownian motion. It's a central object of study in various areas of mathematics, physics, and finance, especially in the Black-Scholes model for option pricing.




# Relation between random walk and white Gaussian noise



The relationship between Random Walk and White Gaussian Noise (WGN) is quite fundamental in stochastic processes and time series analysis. Here’s an exploration of their connection:

1. **Derivation:**
   - A random walk can be derived from a cumulative sum of independent and identically distributed (i.i.d.) random variables, which in a continuous setting could be represented by white noise. 
   - Conversely, the difference between successive values of a random walk results in a sequence of i.i.d. variables, which resembles white Gaussian noise if the underlying variables are normally distributed.

2. **Integration/Differentiation Relationship**:
   - When you take the difference of successive values in a random walk, you obtain a series that resembles white noise. This operation is analogous to differentiation.
   - Conversely, if you take a cumulative sum of a white noise series (or integrate it in a continuous setting), you obtain a random walk.

4. **Statistical Properties**:
   - The increments of a random walk are uncorrelated, which is a property shared with white Gaussian noise.
   - However, while the values of white noise are stationary (the statistical properties do not change over time), the values of a random walk are non-stationary as they depend on the time.
