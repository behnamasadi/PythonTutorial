
# Spectral Analysis
Spectral analysis is a method used in signal processing and statistics to analyze the frequency spectrum of a signal. It breaks down a signal into its constituent sinusoids of different frequencies. Here are key aspects of spectral analysis:

1. **Frequency Decomposition**:
   - Spectral analysis decomposes a signal into a sum of sinusoidal functions, each with a specific frequency, phase, and amplitude. This decomposition can provide insights into the harmonic content of the signal.

2. **Power Spectral Density (PSD)**:
   - One of the main tools in spectral analysis is the Power Spectral Density (PSD), which describes how the power of a signal is distributed over different frequencies. The PSD can reveal dominant frequency components and is crucial for analyzing and interpreting signals in the frequency domain.

3. **Fourier Transform**:
   - The Fourier Transform is a mathematical tool used in spectral analysis to transform a time-domain signal into its frequency-domain representation. The result is a spectrum that shows the amplitude (and phase) of the sinusoidal components as a function of frequency.

4. **Windowing**:
   - In practical applications, windowing is often applied to manage the finite length of data and mitigate spectral leakage. It involves multiplying the signal by a window function before computing the Fourier Transform.

5. **Applications**:
   - Spectral analysis is widely used across many fields including physics, engineering, music, biology, and economics. For example, in electrical engineering, it's used to analyze signals and in economics, it's used for time series analysis.

6. **Software Tools**:
   - There are various software tools and libraries, such as MATLAB and Pythonâ€™s Scipy library, which provide functions for performing spectral analysis.

7. **Periodogram**:
   - The periodogram is an estimate of the spectral density of a signal, often used in spectral analysis to identify the periodic components of a signal.

8. **Comparison and Analysis**:
   - By comparing the spectral content of different signals or the same signal under different conditions, analysts can identify patterns, anomalies, or the effects of particular interventions.

9. **Filtering**:
   - Spectral analysis is also used in the design and analysis of filters, which can remove or enhance certain frequency components of a signal.

Spectral analysis is a powerful technique for understanding the frequency content of signals, and it provides a bridge between time-domain and frequency-domain analysis methods. Through spectral analysis, researchers and practitioners can gain valuable insights into the characteristics and behaviors of various kinds of data and signals.

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate a signal
t = np.linspace(0, 1, 1000, False)  # 1 second
sig = np.sin(2*np.pi*7*t) + np.sin(2*np.pi*13*t)
# Plot signal
plt.figure()
plt.plot(t, sig)
plt.title('Signal')
plt.xlabel('Time [s]')
plt.ylabel('Amplitude')
plt.show()
```

Power Spectral Density (PSD):

```python
from scipy.signal import welch

# Compute PSD
f, Pxx = welch(sig, fs=1000)
# Plot PSD
plt.figure()
plt.semilogy(f, Pxx)
plt.title('PSD')
plt.xlabel('Frequency [Hz]')
plt.ylabel('Power/Frequency [V**2/Hz]')
plt.show()
```

Fourier Transform:

```python
# Compute Fourier Transform
fft_result = np.fft.fft(sig)
frequencies = np.fft.fftfreq(len(fft_result), 1/1000)  # Assuming sample rate of 1000 Hz

# Plot Fourier Transform
plt.figure()
plt.plot(frequencies, np.abs(fft_result))
plt.title('Fourier Transform')
plt.xlabel('Frequency [Hz]')
plt.ylabel('Amplitude')
plt.show()
```

Windowing:

```python
# Apply a Hanning window
windowed_sig = np.hanning(len(sig)) * sig

# Plot windowed signal
plt.figure()
plt.plot(t, windowed_sig)
plt.title('Windowed Signal')
plt.xlabel('Time [s]')
plt.ylabel('Amplitude')
plt.show()
```

Periodogram:

```python
from scipy.signal import periodogram

# Compute periodogram
f, Pxx = periodogram(sig, fs=1000)

# Plot periodogram
plt.figure()
plt.semilogy(f, Pxx)
plt.title('Periodogram')
plt.xlabel('Frequency [Hz]')
plt.ylabel('Power/Frequency [V**2/Hz]')
plt.show()
```
